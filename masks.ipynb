```python
import osmnx as ox
# %matplotlib inline
G_bike = ox.graph_from_place('Amsterdam, NL', network_type='bike')
G_road = ox.graph_from_place('Amsterdam, NL', network_type='drive')
```

```python

import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
import networkx as nx

# Merge the two graphs, with different colors
# Plot the graph
fig, ax = ox.plot_graph(G_bike, edge_color='r', node_size=0, edge_linewidth=0.5, bgcolor='black', show=False, close=False)
fig, ax = ox.plot_graph(G_road, edge_color='b', node_size=0, edge_linewidth=0.2, ax=ax, bgcolor='black', show=False, close=False)

# Save the plot
plt.savefig('amsterdam_mask.png')
```
